```python
from datasets import load_dataset

data = load_dataset("beki/privy",split='train')
```

```python
import pandas as pd
train=pd.DataFrame(data)
```

```python
ner=pd.read_csv('/content/ner_tags.csv')
```

```python
import ast
for i in range(len(ner)):
  ner['tags'][i] = ast.literal_eval(ner['tags'][i])
```

```python
type(ner['tags'][0][5])
```

```python
for i in range(len(train)):
  for j in range(len(train.loc[i]['tags'])):
    if train['tags'][i][j]==ner['tags'][i][j]:
      train['tags'][i][j]=train['tags'][i][j]
    else:
      train['tags'][i][j]=ner['tags'][i][j]
```

```python
tags = data.features['tags'].feature

index2tag = {idx:tag for idx, tag in enumerate(tags.names)}
tag2index = {tag:idx for idx, tag in enumerate(tags.names)}
lst=[]
for i in index2tag.keys():
  lst.append(index2tag[i])
```
